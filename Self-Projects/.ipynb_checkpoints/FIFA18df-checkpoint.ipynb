```python
import pandas as pd


df = pd.read_csv('FIFA18.csv')

club_dict = {}
club_list = []
for item in df['Club']:
    if item not in club_list:
        club_list.append(item)
        
for item in club_list:
    club_dict[item] = club_list.index(item)

for item in df['Club']:
    df['Club_Index'] = df['Club'].map(club_dict)
        
```

```python
df.to_csv('FIFA18_augmented.csv')
```
